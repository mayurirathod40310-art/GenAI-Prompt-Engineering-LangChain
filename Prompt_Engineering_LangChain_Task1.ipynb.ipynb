Task 5

```python
# Role and topic
role = "teacher"
topic = "Neural Networks"

# Define system behavior based on role
system_template = "You are a {role}. Explain the topic clearly."

# User input
user_template = "Explain {topic}"

# Create chat prompt
chat_prompt = ChatPromptTemplate.from_messages([
    ("system", system_template),
    ("user", user_template)
])

# Format it
messages = chat_prompt.format_messages(role=role, topic=topic)

# Print messages
for msg in messages:
    print(msg)
```

Task 6

```python
def validate_inputs(audience, tone):
    
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]
    
    # Check audience
    if audience not in valid_audience:
        print("Invalid audience! Setting default = beginner")
        audience = "beginner"
    
    # Check tone
    if tone not in valid_tone:
        print("Invalid tone! Setting default = formal")
        tone = "formal"
    
    return audience, tone


# Test case
aud, tn = validate_inputs("kids", "friendly")

print(aud, tn)
```

```python
from langchain_core.prompts import PromptTemplate

def generate_prompt(topic, audience, tone, style):
    
    # Step 1: Validate inputs
    audience, tone = validate_inputs(audience, tone)
    
    # Step 2: Select template based on style
    if style == "teaching":
        template = "Explain {topic} for {audience} in a {tone} teaching style"
        
    elif style == "interview":
        template = "Ask interview questions about {topic} for {audience} in a {tone} tone"
        
    elif style == "storytelling":
        template = "Explain {topic} for {audience} in a {tone} storytelling style"
        
    else:
        template = "Explain {topic} for {audience} in a {tone} tone"
    
    # Step 3: Create PromptTemplate
    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )
    
    # Step 4: Generate output
    return prompt.format(topic=topic, audience=audience, tone=tone)


# Test
result = generate_prompt(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)

print(result)
```

```python
from langchain_core.prompts import PromptTemplate

# ONE reusable template
template = "Explain {topic} for {audience} in a {tone} tone"

prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

# 5 different inputs
cases = [
    {"topic": "AI", "audience": "beginner", "tone": "formal"},
    {"topic": "Python", "audience": "intermediate", "tone": "casual"},
    {"topic": "Data Science", "audience": "expert", "tone": "formal"},
    {"topic": "Machine Learning", "audience": "beginner", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "intermediate", "tone": "casual"}
]

# Run all
for case in cases:
    result = prompt.format(**case)
    print(result)
```
